Merge Two Sorted Linked Lists

https://www.codingninjas.com/codestudio/problems/800332?topList=striver-sde-sheet-problems&utm_source=striver&utm_medium=website&leftPanelTab=0

```python
from math import *
from collections import *
from sys import *
from os import *

import sys
from sys import stdin

# Following is the linked list node structure:
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkendList:
    
    def __init__(self):
        self.head = None
        self.tail = None
    
      
def sortTwoLists(first, second):
    # Write your code here.
    
    #print("here") 
    
    answer = LinkendList() 
    curr1 = first 
    curr2 = second 

    
    while curr1 != None and curr2 != None:

        
        if curr1.data <= curr2.data:
            answer.head , answer.tail = insert_tail(curr1.data , answer.head , answer.tail)
            curr1 = curr1.next 
        else:
            answer.head , answer.tail = insert_tail(curr2.data , answer.head , answer.tail)
            curr2 = curr2.next 
    
    while curr1 != None:
        answer.head , answer.tail = insert_tail(curr1.data , answer.head , answer.tail)
        curr1 = curr1.next  

    while curr2 != None:
        answer.head , answer.tail = insert_tail(curr2.data , answer.head , answer.tail)
        curr2 = curr2.next     
        
    return answer.head
            

def insert_tail(data,loc_head,loc_tail):
        
    new_node = Node(data)
        
    if loc_tail == None:
        loc_head = loc_tail = new_node
        
    else:
        loc_tail.next = new_node
        loc_tail = new_node  
    
    return loc_head , loc_tail     
                         
            










def ll(arr):
    
    if len(arr)==0:
        return None
    
    head = Node(arr[0])
    last = head
    
    for data in arr[1:]:
        
        last.next = Node(data)
        last = last.next
        
    return head

def printll(head):
    
    while head:
        
        print(head.data, end=' ')
        
        head = head.next
        
    print(-1)

    

t = int(sys.stdin.readline().strip())

for i in range(t):
    
    arr1=list(map(int, sys.stdin.readline().strip().split(" ")))
    arr2=list(map(int, sys.stdin.readline().strip().split(" ")))
    
    l1 = ll(arr1[:-1])
    l2 = ll(arr2[:-1])
    
    l = sortTwoLists(l1, l2)
    
    printll(l)

```
