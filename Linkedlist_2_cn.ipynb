Intersection of Two Linked Lists

https://www.codingninjas.com/codestudio/problems/630457?topList=striver-sde-sheet-problems&utm_source=striver&utm_medium=website&leftPanelTab=0

```python
from os import *
from sys import *
from collections import *
from math import *


from sys import stdin

#Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None






def findIntersection(firstHead, secondHead) :

	# write your code
    size_first = 0 
    size_second = 0 

    curr1 = firstHead
    curr2 = secondHead

    while curr1 != None:
        size_first = size_first + 1
        curr1 = curr1.next 
    
    while curr2 is not None:
        size_second =  size_second + 1
        curr2 = curr2.next 

    curr1 = firstHead
    curr2 = secondHead   

    if size_first > size_second:

        d= size_first - size_second
        for i in range(0,d):
            curr1 = curr1.next  


    elif size_second > size_first:
        d= size_second - size_first

        for i in range(0,d):
            curr2 = curr2.next 

    ## now both curr1 and curr2 will move same distance

    while curr1 != None:

        if curr1 == curr2:
            return curr1.data
        
        curr1 = curr1.next 
        curr2 = curr2.next 
    
    return -1 


  




#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1
    return head

def attach(head1 , head2) :
	temp1 = head1  

	while(temp1!= None and temp1.next != None) :
		temp1 = temp1.next  
	
	temp2 = head2  
	prev = None 
	
	
	while(temp2.next != None) :
		prev = temp2  
		temp2 = temp2.next  
	
	if(prev == None) :
		if(temp1.data == head2.data) :
			temp1.next = head2  
	
	else :
		if(temp2.data == temp1.data) :
			prev.next = temp1  
	
#main

head1 = takeInput()
head2 = takeInput()
head3 = takeInput()

attach(head1, head2)
#Create Intersection
temp1 = head1  
while(temp1.next != None) :
	temp1 = temp1.next  

temp1.next = head3  

ans = findIntersection(head1, head2)  

print(ans)
```

Add Two Numbers As Linked Lists

```python
from os import *
from sys import *
from collections import *
from math import *

# List Node Class.
class Node:
    def __init__(self, data):

        self.data = data
        self.next = None

    def __del__(self):
        if(self.next):
            del self.next

class LinkedList:

    def __init__(self):
        self.head = None
        self.tail = None

 

def addTwoNumbers(head1, head2):
    # Write your code here.
    curr1 = head1 
    curr2 = head2 

    carry_flag = 0 

    answer = LinkedList()


    while curr1 != None and curr2 != None:

        val, carry_flag = return_val_carry(curr1.data, curr2.data, carry_flag)

        #print(val)
        new_node = Node(val) 
        answer.head , answer.tail = insert_tail(new_node , answer.head , answer.tail)
        #carry_flag = 0

        curr1 = curr1.next
        curr2 = curr2.next 
    
    while curr1 != None:

        val, carry_flag = return_val_carry(curr1.data,0, carry_flag)
        
        new_node = Node(val)
        
        answer.head , answer.tail = insert_tail(new_node , answer.head , answer.tail)
        #carry_flag = 0 
        curr1 = curr1.next

    while curr2 != None:

        val, carry_flag = return_val_carry(0, curr2.data, carry_flag)
        
        new_node = Node(val)
        answer.head , answer.tail = insert_tail(new_node , answer.head , answer.tail)
        #carry_flag = 0 
        curr2 = curr2.next 

    if carry_flag == 1:

        new_node = Node(carry_flag)
        answer.head , answer.tail = insert_tail(new_node , answer.head , answer.tail)         

    return answer.head  

def return_val_carry(data_1 = 0,data_2 = 0, carry = 0):

    val = data_1 + data_2 + carry
    #print(val >= 10)

    if val >= 10 :
        val = val - 10
        #print(val)
        carry = 1
    else:
        carry = 0   

    return val , carry    
        
def insert_tail(new_node ,head,tail):

    if head == None:
        head = tail = new_node
        return head, tail

    else:

        tail.next = new_node
        tail = new_node
        return head , tail 











        


```
